```python
grouped = df.groupby(['Մարզ', 'Օգտվում ե՞ք անկանխիկ վճարամիջոցներից ']).size().unstack(fill_value=0)
grouped_percentage = grouped.div(grouped.sum(axis=1), axis=0) * 100
yes_percent = grouped_percentage["Այո"].reset_index().rename(columns={"Այո": "Տոկոս"})

yes_percent = yes_percent.sort_values("Տոկոս", ascending=True)

fig = px.bar(
    yes_percent,
    x="Տոկոս",
    y="Մարզ",
    orientation='h',
    title="Ո՞ր մարզերում են առավել շատ օգտագործում անկանխիկ վճարամիջոցներ",
    text="Տոկոս",
    color="Տոկոս",
    color_continuous_scale="Blues"
)

fig.update_layout(
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(family="Arial", size=14),
    title_font=dict(size=20, color="#333"),
    xaxis_title="Օգտագործման տոկոսային բաժին (%)",
    yaxis_title="Մարզ",
    margin=dict(l=100, r=50, t=80, b=40),
    coloraxis_showscale=False
)

fig.update_traces(
    texttemplate='%{text:.1f}%',
    textposition='outside',
    marker_line_color='black',
    marker_line_width=1.2
)

fig.show()
```

* Yerevan stands out significantly with the highest digital payment usage, likely due to better infrastructure, higher urbanization, and access to services.
*Rural or less urbanized regions such as Tavush and Aragatsotn show lower usage, possibly indicating limited access or lower digital literacy.

# **Summary**

* **Regional Trends:** Digital tool usage is highest in Yerevan (66.4%), while rural regions like Tavush and Aragatsotn show significantly lower adoption rates, indicating a clear urban–rural gap.

* **Employment Status:** Salaried employees are the most active users of digital financial tools. Usage drops among the self-employed and is lowest among the unemployed.

* **Education Level:** There is a strong positive correlation between education and digital tool usage. Individuals with higher education use them far more than those with only secondary or incomplete education.

* **Income Level:** Higher-income individuals are slightly more likely to use digital tools, though adoption exists across all income levels with a modest difference.

* **Payment Purposes:** Utility payments and purchases are the most common categories for wallet (non-cash) usage, highlighting their role in everyday essential spending.
